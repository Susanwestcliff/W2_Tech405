```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/dataset_w1/TSLA_stock_data.csv')

#dropping datec column
data = data.drop(columns=['Date'])

# Display summary statistics
data.describe()

# Plot pairwise relationships
sns.pairplot(data)
plt.show()
```

```python
# Separating features and target variable
X_features = data.drop('Volume', axis=1)
y_target = np.log(data['Volume'])

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_features, y_target, test_size=0.3, random_state=42)

# Display shapes of training and testing sets
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Scaling features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Linear Regression Model
linear_model = LinearRegression()
linear_model.fit(X_train_scaled, y_train)

# Making predictions and evaluating the model
y_predicted = linear_model.predict(X_test_scaled)
mse_value = mean_squared_error(y_test, y_predicted)
r2_value = r2_score(y_test, y_predicted)

# Plotting actual vs. predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_predicted, alpha=0.6)
plt.xlabel("Actual Volume")
plt.ylabel("Predicted Volume")
plt.title("Susan: Linear Regression: Actual vs. Predicted Stock Volume")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'r--')
plt.show()
```

```python
# Report for Linear Regression
y_predicted = linear_model.predict(X_test_scaled)
mse_value = mean_squared_error(y_test, y_predicted)
r2_value = r2_score(y_test, y_predicted)

print("Linear Regression - Mean Squared Error:", mse_value)
print("Linear Regression - R-squared:", r2_value)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/dataset_w1/TSLA_stock_data.csv')

#dropping datec column
data = data.drop(columns=['Date'])

# Display summary statistics
data.describe()

# Plot pairwise relationships
sns.pairplot(data)
plt.show()
# Separating features and target variable
X_features = data.drop('Volume', axis=1)
y_target = np.log(data['Volume'])

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_features, y_target, test_size=0.3, random_state=42)

# Display shapes of training and testing sets
X_train.shape, X_test.shape, y_train.shape, y_test.shape

# Scaling features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Linear Regression Model
linear_model = LinearRegression()
linear_model.fit(X_train_scaled, y_train)

# Making predictions and evaluating the model
y_predicted = linear_model.predict(X_test_scaled)
mse_value = mean_squared_error(y_test, y_predicted)
r2_value = r2_score(y_test, y_predicted)

# Plotting actual vs. predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_predicted, alpha=0.6)
plt.xlabel("Actual Volume")
plt.ylabel("Predicted Volume")
plt.title("Susan: Linear Regression: Actual vs. Predicted Stock Volume")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'r--')
plt.show()
# Report for Linear Regression
y_predicted = linear_model.predict(X_test_scaled)
mse_value = mean_squared_error(y_test, y_predicted)
r2_value = r2_score(y_test, y_predicted)

print("Linear Regression - Mean Squared Error:", mse_value)
print("Linear Regression - R-squared:", r2_value)
```
